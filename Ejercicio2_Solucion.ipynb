```python
from sklearn import cluster
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
land_water = cv2.imread("data/land_water.jpeg")
```

```python
plt.imshow(cv2.cvtColor(land_water,cv2.COLOR_BGR2RGB)) #Imagen 1
```

```python
vectorized = land_water.reshape((-1,3))
```

```python
vectorized.shape
```

```python
vectorized = np.float32(vectorized)
```

```python
k_means = cluster.KMeans(n_clusters=3) #Clusterizar
_ = k_means.fit(vectorized)
```

```python
X_clustered = k_means.labels_ #Extraer los labels
```

```python
X_clustered.shape
```

```python
X_clustered
```

```python
X_clustered = X_clustered.reshape(land_water.shape[:2])
```
